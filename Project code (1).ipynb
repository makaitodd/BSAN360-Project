```python
daily = (df_project.groupby(["date_occured", "area_name"]).size().reset_index(name="daily_crimes"))
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols("daily_crimes ~ C(area_name)", data=daily).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
central = daily[daily["area_name"] == "Central"]["daily_crimes"]
areas = daily["area_name"].unique()
others = [a for a in areas if a != "Central"]
```

```python
from scipy.stats import ttest_ind
results = []

for area in others:
    other_group = daily[daily["area_name"] == area]["daily_crimes"]
    tstat, pval = ttest_ind(central, other_group, equal_var=False)
    results.append([area, tstat, pval])

t_results = pd.DataFrame(results, columns=["area", "t_stat", "p_value"])
t_results
```

```python
import statsmodels.api as sm
demo_counts = (df_project.dropna(subset=['vict_age', 'vict_sex', 'vict_descent']).groupby(['vict_age', 'vict_sex', 'vict_descent']).size().reset_index(name='crime_count'))
demo_dummies = pd.get_dummies(demo_counts,columns=['vict_sex', 'vict_descent'],drop_first=True)
import statsmodels.api as sm

y = demo_dummies['crime_count']

X = demo_dummies.drop(columns=['crime_count'])
X = X.select_dtypes(include=['number'])
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```
